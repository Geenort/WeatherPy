### Analysis
* As expected, measured temperatures are higher near the equator (0 degrees latitude) and lower near the poles (towards +/- 90 degrees latitude). Furthermore, cities in the southern hemisphere are warmer than cities in the northern hemisphere at this time, when comparing cities in both hemispheres at opposite degrees of latitude. This may be due to the tilt of the earth.
* As latitude approaches 0, humidity seems to increase. However this relationship does not seem to hold outside of -20 to 20 degrees latitude.
* Wind speeds seem to be lower near the equator and higher towards the poles, but the relationship is not very strong.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
countries = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1])
    
    # If the city is unique, then add it to a our cities list
    if city.city_name not in cities:
        cities.append(city.city_name)
        countries.append(city.country_code)

# Print the city count to confirm sufficient count
print(len(cities))
print(len(countries))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# create dataframe with only cities entered in first column. initialize other fields to np.nan so dropna() works later
weather_df = pd.DataFrame(cities)
weather_df.columns = ['City']

weather_df['Cloudiness'] = np.nan
weather_df['Country'] = np.nan
weather_df['Date'] = np.nan
weather_df['Humidity'] = np.nan
weather_df['Lat'] = np.nan
weather_df['Lng'] = np.nan
weather_df['Max Temp'] = np.nan
weather_df['Wind Speed'] = np.nan
```

```python
base_url = f"http://api.openweathermap.org/data/2.5/weather?appid={api_key}&units=metric&q="

print("Beginning Data Retrieval")
print("--------------------------------")

# counter for print log
record_counter = 0

# for each row in DF - use city for query url, increment counter, fill in rest of DF fields
for index, row in weather_df.iterrows():
    
    try:
        response = requests.get(base_url + row['City']).json()
        # KeyError will throw here if no response from API - counter would not increment
        
        record_counter += 1
        
        print(f"Processing Record {record_counter} | {row['City']}")
        
        weather_df.loc[index, 'Cloudiness'] = response['clouds']['all']
        weather_df.loc[index, 'Country'] = response['sys']['country']
        weather_df.loc[index, 'Date'] = response['dt']
        weather_df.loc[index, 'Humidity'] = response['main']['humidity']
        weather_df.loc[index, 'Lat'] = response['coord']['lat']
        weather_df.loc[index, 'Lng'] = response['coord']['lon']
        weather_df.loc[index, 'Max Temp'] = response['main']['temp_max']
        weather_df.loc[index, 'Wind Speed'] = response['wind']['speed']
        
    except KeyError:
        print(f"City of '{row['City']}' not found. Skipping...")
    
print("--------------------------------")
print("Data Retrieval Complete")
print("--------------------------------")
```

```python
weather_df.dropna(inplace=True)
weather_df.count()
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df.to_csv(output_data_file)
weather_df.head()
```
